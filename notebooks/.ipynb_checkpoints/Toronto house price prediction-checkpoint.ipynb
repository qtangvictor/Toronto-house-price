```python
# load link data
df_link = pd.read_csv('../input/post_link.csv')
series_link=df_link['link']
link=series_link.tolist()
# create a dataframe to store listing data
attribute_1=['id','soldPrice','price','bedrooms','bedroomsPartial','bathrooms','bathroomsPartial','unitNumber','neighbourhoodName','city','province','postalCode','type','levels','locker','parking','maintenanceFees','taxes','exterior','basement','garage','driveway','pool','heat','ac','heatingFuel','openHouses','lotFrontage','lotDepth','style','styleName','description']
attribute_misc=['approxAge','balcony','buildingInsuranceIncluded','cableTvIncluded','cacIncluded','commonElementIncluded','condoExp','condoTaxesIncluded','ensuiteLaundry','fireplaceStove','garageParkSpaces','heatIncluded','hydroIncluded','kitchens','parkingFac','parkingIncluded','parkingSpaces','parkingType','stories','waterIncluded','bldgAmenities1Out','bldgAmenities2Out','bldgAmenities3Out','bldgAmenities4Out','bldgAmenities5Out','bldgAmenities6Out']
attribute_other=['link','sqft_min','sqft_max','full_address','latitude','longitude','rooms_totalarea']
df_data = pd.DataFrame(columns=attribute_1+attribute_misc+attribute_other)
row = 0
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
for url in link: # visit each link to extract listing data
    req = urllib.request.Request(url)
    req.add_header('User-Agent',user_agent)
    try:
        time.sleep(1)
        response = urllib.request.urlopen(req,timeout=30).read().decode('utf-8')
        soup = BeautifulSoup(response, 'html.parser')
        json_data = soup.find(id="__NEXT_DATA__").string # find the listing info
        data = json.loads(json_data) # trasform json data to dict   
        data = data['props']['pageProps']['props']['listingData']
    except Exception as e:
        print('link %s generate error: %s' %(url,e))
        continue
    
    for attr in attribute_1:
        try:
            df_data.loc[row,attr] = data[attr]
        except Exception as e:
            print('Index = %d attribute %s generate error: %s' %(row,attr,e))
            
    for attr in attribute_misc:
        try:
            df_data.loc[row,attr] = data['misc'][attr]
        except Exception as e:
            print('Index = %d attribute %s generate error: %s' %(row,attr,e))
            
    if data['rooms']:
        res=0
        for i in data['rooms']:
            if i['width'] and i['length']:
                assert i['unit']=='metres','The unit is '+i['unit']
                res+=i['width']*i['length']*10.7639 # square meter to square ft
        df_data.loc[row,'rooms_totalarea'] = res        
    if data['squareFootage']:
        df_data.loc[row,'sqft_min'] = data['squareFootage']['min']
        df_data.loc[row,'sqft_max'] = data['squareFootage']['max']
    if data['streetNumber'] and data['streetName']:
        df_data.loc[row,'full_address'] = data['streetNumber']+' '+data['streetName']
    if data['position'] and data['position']['coordinates']:
        df_data.loc[row,'latitude'] = data['position']['coordinates'][1]
        df_data.loc[row,'longitude'] = data['position']['coordinates'][0]
    df_data.loc[row,'link'] = url  
    row+=1
df_data.to_csv('house_data.csv',index=False) # write house data to csv file
```

To calculate the distance between the house to the nearest subway station, I downloaded a file containing subway station name,latitude and longitude of the stations in Toronto, storing in subway_stations.csv

```python
# read subway file
# aws
# bucket = 'toronto-house-price-project'
# file_name = 'subway_stations.csv'
# s3 = boto3.resource('s3')
# obj = s3.Object(bucket, file_name)
# subway = pd.read_csv(obj.get()['Body']) 
# aws
subway = pd.read_csv('subway_stations.csv')
def shortest_distance(lat,lng,subway):
    subway_coor = subway[['latitude','longitude']]   
    shortest = float('inf')
    AVG_EARTH_RADIUS = 6371  # in km
    for row in range(subway_coor.shape[0]):
        lat1,lng1 = lat,lng
        lat2,lng2 = subway_coor.loc[row,:]
        lat1, lng1, lat2, lng2 = map(np.radians, (lat1, lng1, lat2, lng2))       
        lat3 = lat2 - lat1
        lng3 = lng2 - lng1
        d = np.sin(lat3 * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lng3 * 0.5) ** 2
        h = 2 * AVG_EARTH_RADIUS * np.arcsin(np.sqrt(d))                    
        shortest = min(shortest,h)
    return shortest
# add a new column called 'subway_distance' into house data
data_house['subway_distance'] = data_house.apply(lambda x:shortest_distance(x.latitude,x.longitude,subway),axis=1) 
```

To get the average household income for neighbourhood in Toronto, I downloaded 2016 census data of Toronto from https://open.toronto.ca/dataset/neighbourhood-profiles/ and geojson file containing the geo information for the 140 neighbourhood in Toronto from https://open.toronto.ca/dataset/neighbourhoods/.

```python
# load the geojson file that defines the boundary of 140 neighbourhoods
bucket = 'toronto-house-price-project'
file_name = 'Neighbourhoods.geojson'
s3 = boto3.resource('s3')
obj = s3.Object(bucket, file_name)
geo_data = json.load(obj.get()['Body']) 
geo_data
```

```python
# find out which neighbourhood the house belongs to
def find_neighbourhood(lat,lng,geodata):
    p = Point(lng,lat)
    for i in range(len(geodata['features'])): #iterate over 140 neighbourhoods
        poly = Polygon(geodata['features'][i]['geometry']['coordinates'][0])
        if p.within(poly):
            return (geodata['features'][i]['properties']['AREA_SHORT_CODE'],geodata['features'][i]['properties']['AREA_NAME'])#return id and name
    return (None,None) # if not within any neighbourhood, return NAN
data_house['neighbourhood_id'] = data_house.apply(lambda x:find_neighbourhood(x.latitude,x.longitude,geo_data)[0],axis=1)
data_house['neighbourhood_name'] = data_house.apply(lambda x:find_neighbourhood(x.latitude,x.longitude,geo_data)[1],axis=1)
data_house.head()
```

There are 1228 nan values for sold price. The average difference between sold price and list price is 4.7%. We can build a simple linear model to predict sold price with list price for those missing sold price

```python
from sklearn.linear_model import Ridge
missing_data = data[np.isnan(data.soldPrice)][['id','price']]
x_train = data[~np.isnan(data.soldPrice)]['price']
y_train = data[~np.isnan(data.soldPrice)]['soldPrice']
x_test = missing_data['price']
l1regr=Ridge()
l1regr.fit(x_train.to_numpy().reshape(-1, 1),y_train)
score_l1regr=l1regr.score(x_train.to_numpy().reshape(-1, 1),y_train)
print('The R square score of the linear model is %f' % score_l1regr)
predict = l1regr.predict(x_test.to_numpy().reshape(-1, 1))
missing_data['soldPrice'] = predict
for index,row in data.iterrows():
    if np.isnan(row['soldPrice']):
        data['soldPrice'][index] = float(missing_data[missing_data.id==row['id']]['soldPrice'])
```

## layout features

 'bedrooms','bedroomsPartial','bathrooms','basement','garage',kitchens,'balcony','condoExp','driveway'

```python
# for basement fill nan and rename
data["basement"] = data["basement"].fillna("missing")
def rename(x):
    if x=='Finished' or x=='None' or x=='Finished and Sep Entrance' or x=='Unfinished' or x=='missing':
        return x
    if 'apartment' in x.lower():
        return 'None'
    if 'unfinished' in x.lower() or 'part' in x.lower():
        return 'Unfinished'
    if 'sep entrance' in x.lower():
        return 'Finished and Sep Entrance'
    if 'finished' in x.lower() or 'fin' in x.lower():
        return 'Finished'  
    return 'Others'
data.basement = data.basement.map(lambda x:rename(x))

```

```python
# create a new category for driveway, condoExp for those with less than 100 counts
data = data.apply(lambda x: x.mask(x.map(x.value_counts())<100, 'Other') if x.name=='driveway' or x.name=='condoExp' else x)

# for garage feature, some values represent the same type but use different names. Rename them
maps = {'Underground':'Undergrnd','Detached Garage':'Detached','No Garage':None,'Attached Garage':'Attached','Shared':'Other','Interlocked':'Other','Detached garage':'Detached','Visitor Parking':'Other','Garage':'Other'}
data['garage'] = data['garage'].replace(maps)

# replace nan with 'missing'
data["garage"] = data["garage"].fillna("missing")
data["balcony"] = data["balcony"].fillna("missing")
data["condoExp"] = data["condoExp"].fillna("missing")
data["driveway"] = data["driveway"].fillna("missing")
```

# training models
I will use 5-fold cross validation and mean absolute percentage error as the metrics

```python
def mape(y_true, y_pred):
    # calculate mean_abs_percentage_error
    error = 0
    # loop over all samples in true and predicted list
    for yt, yp in zip(y_true, y_pred):
        error += np.abs(yt - yp) / yt
    # return mean percentage error
    return error / len(y_true)*100
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import KFold
#from sklearn.model_selection import cross_val_score
xgb_model = xgb.XGBRegressor()
rf_model = RandomForestRegressor()
tree_model = DecisionTreeRegressor()
knn_model = KNeighborsRegressor()
lgb_model = lgb.LGBMRegressor()
gbdt_model = GradientBoostingRegressor()
models=[('KNN', knn_model),('DecisionTree', tree_model),('XGBoost', xgb_model),('RandomForest', rf_model),('lightgbm', lgb_model),('GradientBoost', gbdt_model)]
kf = KFold(n_splits=5,random_state=42,shuffle=True)
mean_error=[]
for name,model in models:
    error=[]
    for train_index, val_index in kf.split(x_train_label):     
        model.fit(x_train_label.loc[train_index],y_train.loc[train_index])
        error.append(mape(y_train.loc[val_index],model.predict(x_train_label.loc[val_index])))
    mean_error.append(np.mean(error))
    
sns.barplot(x=mean_error,y=[i[0] for i in models])        
plt.ylabel('model', fontsize=10)
plt.xlabel('mean absolute percentage error (%)')
plt.show()        
```

XGBoost,Lightgbm,RandomForest have the best performance, achiving a similar error of about 12%. I will choose XGBoost (error is 12.11%) as our prediction model and keep tuning the hyperparameters to optimize the performance.

## Hyperparameters tuning

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
mape_score = make_scorer(mape, greater_is_better=False)
params = {
    "max_depth":[3,7,12], 
    "subsample": [0.6,0.8,1],
    'colsample_bytree':[0.6,0.8,1],
    'eta':[0.01,0.05,0.1],
    'gamma':[0.1,0.5,1]
}
rgr=GridSearchCV(estimator=xgb_model, param_grid=params,scoring=mape_score,cv=5)
rgr.fit(x_train_label,y_train)
```
